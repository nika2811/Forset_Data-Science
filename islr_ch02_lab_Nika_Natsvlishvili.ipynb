```python
x = np.linspace(-np.pi, np.pi, num = 50)
y = x

def f(x,y):
    return np.cos(y)/(1 + np.square(x))

X, Y = np.meshgrid(x,y)
Z = f(X,Y)

fig = plt.figure()
ax = plt.axes(projection = "3d")
ax.contour3D(X,Y,Z,50, cmap = 'binary')
plt.show()
```
