```python
import numpy as np
import matplotlib.pyplot as plt
```

# 1

სიმულირებული `(x1, y1)` წყვილებისთვის დაამოდელირეთ პოლინომური რეგრესიები და შეადარეთ ცდომილებები.




```python
num_dots = 200

np.random.seed(42)

x1 = np.random.rand(num_dots) * 6 - 3
y1 = np.zeros(num_dots)

for i, p in enumerate([-3, 0, 3, 1]):
    y1 = y1 + np.power(x1, i) * p

y1 = y1 + np.random.normal(loc=0, scale=5, size=num_dots)
plt.scatter(x1, y1)
plt.show()
```

```python
def reg_model(x, y, deg=1):
    reg_poly = np.polyfit(x, y, deg)
    reg_poly = np.poly1d(reg_poly)
    return reg_poly
R = {}
for i in range(1,25):
    
    poly = reg_model(x1, y1, i)
    yhat = poly(x1)
    plt.scatter(x1, y1)
    plt.scatter(x1, yhat)
    plt.show()
    
    rss = np.power(y1 - yhat, 2).sum()
    tss = np.power(y1 - y1.mean(), 2).sum()
    r_squared = 1 - rss / tss
    r_adjusted = 1 - (1 - r_squared)*(num_dots - 1)/(num_dots - i - 1 - 1)
    R[f'{i} order polynomial '] = r_adjusted
    print(f"{i} order polynomial has R-squared adjusted={r_adjusted}")
    
max_key = max(R, key=R.get)
print(f"{max_key} has a high value of R")
```

$$
\begin{cases}
x+2y=11\\
5x-3y=3
\end{cases}
$$


### 1)

```python
list = [[1,2],[5,-3]]
A = np.array(list)
```

```python
inv_A = np.linalg.inv(A)
print(inv_A)

B = np.array([11, 3])
X = np.linalg.inv(A).dot(B)

print(X)
```
